```julia
iris_data = CSV.read("data/iris.data",copycols=true,
    header=["sep_len","sep_width","pet_length","pet_width","species"])
last(iris_data,5)
```

```julia
@time begin
using Query
using DataFrames

iris_data = @from i in iris_data begin
    @where i.sep_width>=0. # && i.children > 2
    @select {i...}
    @collect DataFrame
end

last(iris_data,6)
end
```
